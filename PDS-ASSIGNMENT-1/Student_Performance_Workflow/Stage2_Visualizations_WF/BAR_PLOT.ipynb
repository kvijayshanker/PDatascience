```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load the  data into varaible
df = pd.read_csv("C:\\Users\\Vijay Shanker\\Downloads\\StudentsPerformance.csv")
```

```python
# Bar Plot for the Gender Distribution i.e Count vs Gender:

gender_count = df['gender'].value_counts()
plt.figure(figsize=(6, 4))
gender_count.plot(kind='bar', color='skyblue')
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()

plt.savefig("bar_gender.png")
```
