```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load the  data into varaible
df = pd.read_csv("C:\\Users\\Vijay Shanker\\Downloads\\StudentsPerformance.csv")
```

```python
# Box Plot for the Box Plot of Math, Reading, and Writing Scores:
plt.figure(figsize=(8, 6))
df[['math score', 'reading score', 'writing score']].boxplot()
plt.title('Box Plot of Scores')
plt.ylabel('Score')
plt.show()

plt.savefig("BOX_MWR.png")
```
