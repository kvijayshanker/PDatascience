```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load the  data into varaible
df = pd.read_csv("C:\\Users\\Vijay Shanker\\Downloads\\StudentsPerformance.csv")
```

```python
# Pie Chart for the Test Preparation Course Distribution :

lunch_count = df['lunch'].value_counts()
plt.figure(figsize=(6, 6))
plt.pie(lunch_count, labels=lunch_count.index, autopct='%1.1f%%', colors=['gold', 'lightblue'])
plt.title('Lunch Type Distribution')
plt.show()
```
