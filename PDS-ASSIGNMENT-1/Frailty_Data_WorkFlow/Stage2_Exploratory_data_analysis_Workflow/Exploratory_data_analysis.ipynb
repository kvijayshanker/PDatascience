```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python

# Load the clean data into varaible
data = pd.read_csv("C:\\Users\\Vijay Shanker\\Downloads\\pds files\\clean_data.csv")
```

```python
#upload the data into dataframe
df = pd.DataFrame(data)
```

```python
#Scatter plot of "Grip strength" vs. "Weight (kg)":
plt.figure(figsize=(8, 6))
plt.scatter(df['Grip strength'], df['Weight (kg)'], c='blue', label='Data points')
plt.title('Grip Strength vs. Weight (kg)')
plt.xlabel('Grip Strength')
plt.ylabel('Weight (kg)')
plt.legend()
plt.show()
```

```python
# Describe the clean data in form of stats
print(data.describe())
```

```python
#Box plot of "Grip strength" vs "Frailty":

plt.figure(figsize=(8, 6))
sns.boxplot(x='Frailty', y='Grip strength', data=df, palette='Set2')
plt.title('Grip Strength by Frailty')
plt.xlabel('Frailty (Y/N)')
plt.ylabel('Grip Strength')
plt.show()
```
